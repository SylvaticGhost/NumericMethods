```python
import numpy as np
import scipy as sp
from main import Program
from copy import copy
np.set_printoptions(suppress=True)
```

#### Виконання програмного розв'язку для отримання данних

```python
program = Program()
λ, xs = program.Solve()
A = copy(program.A)
```

#### Результати програмного розв'язку

```python
print(f"λ = {λ}")
print(f'x = ')
for i in range(len(xs)):
    print(f'x{i+1} = {xs[i]}')
```

#### Розв'язок мат. пакетом

```python
r = sp.linalg.eig(A)
print('Власні значення:')
print(f"λ = {r[0].real}")

print('Власні вектори:')
for i in range(len(r[1].T)):
    vector = r[1].T[i]
    normalized_vector = vector / vector[-1]
    print(f'x{i+1} = {normalized_vector}')


```
